<h1>Turning Difficulty to a Number</h1>

```python
for i in range(len(data)):
    if data.loc[i, "difficulty"] == "Easy":
        data.loc[i, "difficulty"] = 0
    if data.loc[i, "difficulty"] == "Medium":
        data.loc[i, "difficulty"] = 1
    if data.loc[i, "difficulty"] == "Hard":
        data.loc[i, "difficulty"] = 2

for i in range(len(data_not_solved)):
    if data_not_solved.loc[i, "difficulty"] == "Easy":
        data_not_solved.loc[i, "difficulty"] = 0
    if data_not_solved.loc[i, "difficulty"] == "Medium":
        data_not_solved.loc[i, "difficulty"] = 1
    if data_not_solved.loc[i, "difficulty"] == "Hard":
        data_not_solved.loc[i, "difficulty"] = 2
data_not_solved
```

<h1>Finding Average</h1>

```python
avgDiff = 0
avgLike = 0
avgDiss = 0

#Assumption is that, the question similar to the average of the questions I have already solved should be easier for me to solve
for i in range(len(data)):
    avgDiff = avgDiff + data.loc[i, "difficulty"]
    avgLike = avgLike + data.loc[i, "likes"]
    avgDiss = avgDiss + data.loc[i, "disslikes"]

avgDiff = avgDiff / len(data)
avgLike = avgLike / len(data)
avgDiss = avgDiss / len(data)

print(avgLike, avgDiss, avgDiff)
```

<h1>Find the Next Optimal Question</h1>

```python
import sys

optimal = 0
optimalValue = sys.maxsize * 2 + 1#giving it a size that is big enough to mimic Double.MAX_VALUE of Java.
for i in range(len(data_not_solved)):
    #normalize likes and disslikes
    value = abs(data_not_solved.loc[i, "likes"] - avgLike)/(data_not_solved.loc[i, "likes"]+1) + abs(data_not_solved.loc[i, "difficulty"] - avgDiff)/3 + abs(data_not_solved.loc[i, "disslikes"] - avgDiss)/(data_not_solved.loc[i, "disslikes"]+1)
    if value < optimalValue:
        optimalValue = value
        optimal = i

print("The most optimal question is", data_not_solved.loc[optimal, "name"],"with the score of", optimalValue)

#IMPORTANT NOTES
#After solving this question, we can then delete this question from the data_not_solved and add it to data and run all of these again to find the next best question to solve.
#I wanted to add more data to this project like acceptance_rate, tags of the questions like dynamic_programming, string, binary_tree and number of solution posts, number of discussion posts... Especially the tags section because domain of questions change it's solveability a lot.
#At the time of running first(12.28.2022 2:41 AM) it's recommending me to solve "simplify-path" with the score of 102.64251207729467, I have looked at the question it looks enjoyable and solveable.
```
